```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import koreanize_matplotlib
# df = pd.read_csv('../data_resources/서울특별시 공공자전거 이용정보(월별)_24.7-12.csv', encoding='cp949')
```

## 1) 범주형 데이터

```python
plt.figure(figsize=(5, 3))
sns.countplot(data=df, x="성별")
plt.title("성별 분포")
plt.grid()
plt.show()
```

```python
#성별에 따른 만족도를 시각화

plt.figure()
sns.countplot(data=df, x="만족도", palette="pastel", order=['하', '중', '상'])
plt.title("만족도 분포")
plt.grid()
plt.show()
```

```python
#성별에 따른 만족도를 시각화

plt.figure()
sns.countplot(data=df, x="만족도", hue="성별", palette="pastel", order=['하', '중', '상'])
plt.title("만족도 분포")
plt.grid()
plt.show()
```

## 2) 연속형 데이터

```python
# 월 수입(만원) 열에 대한 히스토그램

plt.figure()
plt.hist(df["월 수입(만원)"], bins=15, color='gold', edgecolor="k")
plt.grid()
plt.title("월 수입(만원) 분포 확인")
plt.show()
```

```python
# 월 수입(만원) 열에 대한 히스토그램

plt.figure()
sns.histplot(data=df, x="월 수입(만원)", bins=20, kde=True)
plt.grid()
plt.title("월 수입(만원) 분포 확인")
plt.show()
```

```python
plt.figure()
plt.boxplot(x=df["월 수입(만원)"])
plt.title("월 수입에 대한 Box-Plot")
plt.xlabel("월 수입(단위: 만 원)")
plt.show()
```

```python
plt.figure()
sns.boxplot(data=df, x='성별', y="월 수입(만원)")
plt.title("월 수입에 대한 Box-Plot")
plt.show()
```

## 4) 이산형 데이터

```python
# 방문 페이지 수
fig, ax = plt.subplots()
ax.set_title("방문 페이지 수 분포")
sns.countplot(data=df, x="방문 페이지 수", ax=ax)
plt.show()
```

### -날짜에 대한 데이터 타입은 datetime 이다.

```python
df = pd.read_csv('../data_resources/sample_data.csv', parse_dates=['날짜'])
```

```python
df.info()
```

```python
df.날짜 = pd.to_datetime(df.날짜)
```

```python
df.info()
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.plot(df["날짜"], df["월 수입(만원)"], marker='o')
plt.grid()
ax.set_title("시간에 따른 체류 시간 변화 그래프")
plt.show()
```

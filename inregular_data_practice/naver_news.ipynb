# 네이버 뉴스 API로 가로수길 공실 뉴스들 가져오기

```python
import requests
import os
from dotenv import load_dotenv
load_dotenv()

# 1. 정보 입력
client_id = os.getenv("NAVER_CLIENT_ID")
client_secret = os.getenv("NAVER_CLIENT_SECRET")

# 2. 키워드 입력
keyword = "가로수길 공실"

# 3. URL 만들기
base_url = "https://openapi.naver.com/v1/search/news.json"
headers = {
    "X-Naver-Client-Id": client_id,
    "X-Naver-Client-Secret": client_secret
}
sort = "sim"
## 반복 요청
data_list = []

for start in range(1, 1002, 100):
    display = 100

    if start == 901:
        display = 99
    if start == 1001:
        start -= 1

    params = {
        "query": keyword,
        "display": display,
        "start": start,
        "sort": sort
    }

    # 5. 데이터 요청
    response = requests.get(base_url, headers=headers, params=params)

    if response.status_code == 200:
        response_json = response.json()
        data_list.extend(response_json["items"])
    
    else:
        print("Error")
```

```python

        
print(len(data_list))

```

```python
import pandas as pd

garosugil_gongsil = pd.DataFrame(data_list)
garosugil_gongsil.to_csv("../data_resources/garosugil_gongsil.csv")
```

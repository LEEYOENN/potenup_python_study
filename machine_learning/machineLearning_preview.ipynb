```python
# appreply2.csv
import pandas as pd
data = pd.read_csv("../data_resources/appreply2.csv")
data
```

## 2) 형태소 분석기

```python
from konlpy.tag import Okt
import re

okt = Okt()

# 미션: 각 문장을 정규 표션식으로 전처리한 후, 형태소 분석기를 돌려서 문장별로 담아주세요.

review_list = []

for i, sent in enumerate(data["text"].to_list()) :
    if i % 10 == 0:
        print(i, "번째 리뷰")

    new_review = re.sub(r"[^0-9a-zA-Z가-힣\s]", "", sent)
    result = okt.pos(new_review, norm=True, stem=True)

    words = [ word for word, pos in result ]
    string = " ".join(words)
    

    review_list.append(string)

```

```python
okt = Okt()

# 미션: 각 문장을 정규 표션식으로 전처리한 후, 형태소 분석기를 돌려서 문장별로 담아주세요.

review_list = []

for i, sent in enumerate(data["text"].to_list()) :
    if i % 10 == 0:
        print(i, "번째 리뷰")

    new_review = re.sub(r"[^0-9a-zA-Z가-힣\s]", "", sent)
    result = okt.morphs(new_review, norm=True, stem=True)

    string = " ".join(result)
    

    review_list.append(string)
```

## 3) LDA

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(
    max_df = 0.1,               # 전체 단어 등장 비율이 p 이상인 것만 사용할 것.
    min_df = 2,                 # 단어가 적어도 n개 이상인 것만 사용할 것.
    max_features = 1000,        # 상위 항목 n개만 사용할 것.
    ngram_range= (1,2)          # 단어의 조합
)
```

```python
# n_gram 이란? 
# 나는 오늘 밥을 먹었다
# 나는 오늘 / 오늘 밥을 / 밥을 먹었다 / 2_gram
```

```python
feature_vec = vectorizer.fit_transform(review_list)

feature_vec.shape
```

```python
feature_names = vectorizer.get_feature_names_out()
print(feature_names)
```

## 4) LDA - 문장을 클러스터링

```python
from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(n_components=3, n_jobs=-1)
lda.fit(feature_vec)
```

```python
print(lda.components_.shape)
print(lda.components_)
```

```python
import sklearn
print(sklearn.__version__)
```

```python
import numpy as np
```

```python
lda.components_
```

```python
for idx, topic in enumerate(lda.components_) :
    print(f"# Topic {idx + 1}: ")
    print(topic)
    print(topic.argsort())
    topic_word_idx = topic.argsort()[::-1]
    top_idxs = topic_word_idx[:20]
    topics = [feature_names[idx] for idx in top_idxs]

    print(" ".join(topics))
```

## 5) 문장 살펴보기

```python
sentence_topic = lda.transform(feature_vec)
print(sentence_topic.shape)
print(sentence_topic)
```

```python
sent_list = []

for n in range(sentence_topic.shape[0]):
    topic_idx = sentence_topic[n].argmax()
    topic_prob = sentence_topic[n].max()

    sent_list.append([n, topic_idx, topic_prob])

topic_per_data = pd.DataFrame(sent_list, columns=["No", "topic_No", "prob"])
topic_per_data["review"] = data["text"]

topic_per_data
```

# 2. 분류 학습(지도 학습)

```python
import pandas as pd
data = pd.read_csv("../data_resources/appreply2.csv")
data
```

```python
# score를 기반으로 긍정/부정 나누고싶다.

import numpy as np

data["reaction"] = np.where(data["score"] > 4, 1, 0)

data.head()
```

```python
data['length'] = data['text'].str.len()
```

```python
data
```

```python
import matplotlib.pyplot as plt
import koreanize_matplotlib

plt.figure()
plt.title("review length histogram")
plt.hist(data['length'], bins=50)
plt.show()
```

## 2) 데이터 전처리

```python
import re
okt = Okt()

# 미션: 각 문장을 정규 표션식으로 전처리한 후, 형태소 분석기를 돌려서 문장별로 담아주세요.

review_list = []

for i, review in enumerate(data["text"].to_list()) :
    if i % 10 == 0:
        print(i, "번째 리뷰")

    new_review = re.sub(r"[^0-9a-zA-Z가-힣\s]", "", review)
    result = okt.morphs(new_review, norm=True, stem=True)

    string = " ".join(result)
    

    review_list.append(string)
```

```python
review_list
```

## 3) TFIDFVectorizer

```python
from sklearn.feature_extraction.text import TfidfVectorizer


vectorizer = TfidfVectorizer()

tfidf_vec = vectorizer.fit_transform(review_list)
print(vectorizer.vocabulary_)
print(tfidf_vec.toarray())
```

count_vectorizer => 단어가 문장마다 등장한 횟수를 세어 행렬로 만들어줌

Tfidf_vectorizer => Term Frequency: 한 문서에서 특정단어가 얼마나 등장하는지, IDF: 전체 문서 중에서 특정 단어가 얼마나 희귀한지

## 4) 학습 데이터 준비

```python
from sklearn.model_selection import train_test_split

X = tfidf_vec
Y = data["reaction"]
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

## 5) 데이터 학습

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x_train, y_train)
```

## 6) 성능 평가

```python
print(lr.score(x_test, y_test))
print(lr.coef_.shape)
```

```python
y_pred = lr.predict(x_test)
```

```python
import numpy as np
print(np.unique(y_pred, return_counts=True))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(f"Precision: {precision_score(y_test, y_pred)}")
print(f"F1score: {f1_score(y_test, y_pred)}")
print(f"Recall Score: {recall_score(y_test, y_pred)}")
confusion_matrix(y_test, y_pred)
```
